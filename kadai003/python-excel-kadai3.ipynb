```python
import pandas as pd
import openpyxl
```

```python
# 処理対象のファイル名のリストを格納
file_list = ['2022_年間売上表.xlsx', '2023_年間売上表.xlsx']

# 結合後のデータを格納するための空のデータフレームを作成
df = pd.DataFrame()

# ファイルを一つずつ処理する
for file_name in file_list:
    # Excelファイルを読み込む
    temp_df = pd.read_excel(file_name, header=0)

    # データフレームの列名を指定
    temp_df.columns = ['売上年', '商品', '金額(千円)']

    # concat()でデータフレームを連結
    df = pd.concat([df,temp_df], ignore_index=True)

df.columns=  ['売上年', '商品', '金額(千円)']
```

```python
# 連結したデータを「商品」「売上年」ごとにグループ化し、合計金額に集約
df_gpsum = df.groupby(['商品','売上年']).sum()
```

```python
# 新規Excelファイルを作成する
writer = pd.ExcelWriter('売上集計表.xlsx')

# 作成したDataFrameオブジェクトをExcelファイルに書き込む
df_gpsum.to_excel(writer,sheet_name='Sheet1',index=True)

# Excelファイルを閉じる
writer.close()
```

```python
from openpyxl.styles import PatternFill
from openpyxl.styles.borders import Border, Side
```

```python
# 「売上集計表.xlsx」のヘッダー部分（1行目）を薄いグレー（#F2F2F2）に設定

wb = openpyxl.load_workbook('売上集計表.xlsx')
ws = wb.active


for row1 in ws['A1':'C1']:
    for cell in row1:
        cell.fill = PatternFill(patternType='solid', fgColor='F2F2F2')

side = Side(style='thin', color='000000')
border = Border(top=side, bottom=side, left=side, right=side)

for row2 in ws:
    for cell in row2:
        ws[cell.coordinate].border = border


wb.save('売上集計表.xlsx')
```
